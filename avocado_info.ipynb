```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pylab as plt

print('Setup completo!')
```

```python
# carregando os dados dos abacates e escolhendo a data como o índice
data = pd.read_csv('avocado.csv', parse_dates=['Date'])
```

```python
data.head()
```

```python
# agrupando por data e calculando a média do volume total escolhendo resetar o indice
summr = data.groupby('Date')['Total Volume'].mean().reset_index()
```

```python
summr.head()
```

```python
data.index
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,5))

summr.set_index('Date').plot(ax=ax, marker='o', linestyle='-', color='m')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,10))

summr = summr.set_index('Date')

summr['month'] = summr.index.month

grouped = summr.groupby('month')['Total Volume'].agg(['mean', 'std', 'median', 'min', 'max'])
grouped.plot(ax=ax, marker='D', linewidth=2)
ax.set_label('Mês')

plt.show()
```
